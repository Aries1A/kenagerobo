```python
import pandas as pd
import numpy
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("gym-kenage/data/q_data__.csv")
```

```python
df
```

```python
plt.plot(df["pos_x"],df["pos_y"])
```

```python
df['episode'] [2:]= 0
```

```python
df_index = df[df['step'] == 0].index
df_index
```

```python
# episodeをdfに追加する
for i in range(len(df_index)):
    df['episode'] [df_index[i]:]= i
```

```python
# episodeごとの軌跡を図で出力
for i in range(len(df_index)):
    df_episode = df[df['episode'] == i]
    if len(df_episode) > 30:
        plt.plot(df_episode["pos_x"],df_episode["pos_y"])
        plt.xlim(0,270)
        plt.ylim(0,90)
        plt.show()
```
